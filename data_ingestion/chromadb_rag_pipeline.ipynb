##create sample data

```python
sample_docs = [
    """Machine Learning Fundamentals:
    Machine learning is a subset of artificial intelligence that enables systems to learn and improve from experience without being explicitly programmed.""",

    """Deep Learning and Neural Networks:
    Deep learning is a subset of machine learning based on artificial neural networks with multiple layers that can learn increasingly abstract representations of data.""",

    """Natural Language Processing:
    Natural Language Processing (NLP) is a branch of AI that helps computers understand, interpret, and generate human language."""
]
```

```python
sample_docs
```

```python
#save the sample documents to the files
import tempfile

# Create temporary directory
temp_dir = tempfile.mkdtemp()

# Save sample documents to files
for i, doc in enumerate(sample_docs):
    with open(f"{temp_dir}/doc_{i}.txt", "w") as f:
        f.write(doc)

print(f"Sample documents created in: {temp_dir}")
```

```python
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader(
    temp_dir,
    glob="*.txt",
    loader_cls=TextLoader
)

documents = loader.load()

print(f"Number of documents: {len(documents)}")
print(documents[0].page_content)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(documents)

print(f"Number of chunks: {len(chunks)}")
print(chunks[0].page_content)
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="./chroma_db"
)

query = "What is machine learning?"

results = vectorstore.similarity_search(query, k=2)

for doc in results:
    print(doc.page_content)
    print("-" * 50)
```
